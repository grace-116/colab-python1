```python
# 3행 3열의 2차원 센서 데이터 리스트 생성
sensor_grid = [
    [10, 20, 30],  # 인덱스 0 (첫 번째 행)
    [40, 50, 60],  # 인덱스 1 (두 번째 행)
    [70, 80, 90]   # 인덱스 2 (세 번째 행)
]

# 2행 3열(인덱스 [1][2])의 데이터 접근 및 수정
print("수정 전 값:", sensor_grid[1][2])
sensor_grid[1][2] = 99
print("수정 후 값:", sensor_grid[1][2])


# 1. 톱니형 리스트 (각 행의 길이가 다름)
jagged_list = [
    [1.1, 1.2],
    [2.1, 2.2, 2.3, 2.4],
    [3.1]
]
print("톱니형 리스트의 2번 행 요소 개수:", len(jagged_list[1]))

# 2. 2차원 튜플 (튜플 안에 튜플)
immutable_grid = (
    (10, 20),
    (30, 40)
)
# immutable_grid[0][0] = 99 -> 튜플은 수정 불가
```

```python
data_matrix = [
    [1, 2],
    [3, 4],
    [5, 6]
]

# for문으로 요소 직접 순회
for x in data_matrix:
    for y in x:
        print(y, end=" ")
    print()  # 한 행의 출력이 끝나면 줄바꿈

# for와 range로 인덱스를 생성하여 순회
for i in range(len(data_matrix)):            # 행의 크기만큼 반복
    for j in range(len(data_matrix[i])):     # 각 행의 열 크기만큼 반복
        print(f"[{i}][{j}]위치: {data_matrix[i][j]}", end="  ")
    print()

# while 반복문으로 2차원 리스트 순회
i = 0
while i < len(data_matrix):
    j = 0
    while j < len(data_matrix[i]):
        print(data_matrix[i][j], end=" ")
        j += 1
    print()
    i += 1 #for문과 다르게 while문은 1증가하는 변화식을 꼭 써줘야 한다.
```

```python
# 행 인덱스와 열 인덱스가 동일한 지점(주 대각선 방향)의 요소들만 추출하여 그 합계를 구하는 코드를 작성하시오.
# 'for'와 'range'를 사용하여 인덱스를 제어해야 합니다.

matrix = [
    [10, 20, 30],
    [40, 50, 60],
    [70, 80, 90]
]
diagonal_sum = 0

for i in range(len(matrix)):
    diagonal_sum += matrix[i][i]

print(diagonal_sum)
```

```python
# 1. 1차원 리스트 생성 (0으로 5칸 초기화)
list_1d = []
for i in range(5):
    list_1d.append(0)
print("1차원 리스트:", list_1d)

# 2. 2차원 리스트 생성 (0으로 3행 4열 초기화)
list_2d = []
for i in range(3):
    row = []
         # 임시 1차원 리스트(행) 생성
    for j in range(4):
        row.append(0)     # 행에 0을 4번 추가
    list_2d.append(row)   # 완성된 행을 2차원 리스트에 추가

print("2차원 리스트:", list_2d)
```

```python
# 리스트 표현식으로 3행 4열 독립 행렬 초기화
# 해석: [0, 0, 0, 0]을 생성하는 과정을 3번 반복하여 묶음
comprehension_2d = [[0 for j in range(4)] for i in range(3)]

# 첫 번째 행의 요소 하나만 변경해 독립성 검증
comprehension_2d[0][0] = 99
print("리스트 표현식 결과:", comprehension_2d)


# 각 행이 가져야 할 열의 길이를 정의한 리스트
row_sizes = [3, 1, 4, 2]

# row_sizes의 요소를 size로 받아 그 크기만큼 0을 생성
jagged_comprehension = [[0 for j in range(i)] for i in row_sizes]

for row in jagged_comprehension:
    print(row)


# [장비 ID, 공정 소요 시간, 온도] 로 구성된 2차원 데이터
process_records = [
    ["EQ_C", 12.5, 200],
    ["EQ_A", 15.2, 180],
    ["EQ_B", 10.1, 210]
]

# 1. 기본 정렬 (첫 번째 요소인 장비 ID 알파벳 순)
default_sorted = sorted(process_records)
print("기본 정렬:", default_sorted)

# 2. 특정 인덱스 기준 정렬 (두 번째 요소인 공정 소요 시간 기준)
# lambda x: x[1] 은 각 행(x)에서 인덱스 1번 데이터를 정렬 키로 쓰겠다는 의미
time_sorted = sorted(process_records, key=lambda x: x[2])
print("온도 기준 정렬:", time_sorted)
```
