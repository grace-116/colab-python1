```python
# 암호 해독 및 데이터 조작 프로젝트
cipher_text = '!1n2u3f4 5s6i7 8n9o0h1t2y3P4'

# 1. cipher_text에서 슬라이스 증가폭을 사용하여 숫자만 제외하고 알파벳과 기호만 추출하여 step1 변수에 저장하십시오.
step1 = cipher_text[::2]

# 2. step1 문자열에 음수 증가폭 슬라이스를 적용하여 문장을 거꾸로 뒤집고 step2 변수에 저장하십시오.
step2 = step1[::-1]

# 3. step2 문자열의 첫 글자인 대문자 P를 소문자로 바꾸기 위해 먼저 list 함수를 사용하여 step2를 리스트로 변환하고 step3 변수에 저장하십시오.
step3 =list(step2)

# 4. 인덱싱을 사용하여 step3 리스트의 첫 번째 요소를 소문자 p로 교체하십시오.
# 코드를 작성하십시오.
step3[0] = 'p'

# 5. 데이터가 더 이상 훼손되지 않도록 step3 리스트를 튜플로 변환하여 final_result 변수에 저장하고 출력하십시오.
final_result = tuple(step3)
print(final_result)
```
