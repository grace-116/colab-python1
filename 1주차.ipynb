```python
# Advanced. 논문/보고서용 지수 표기법 (Uncertainty나 아주 작은 확률값)
prob = 0.000000045678
print(f"실패 확률: {prob:.3e}")  # 결과: 4.568e-08 -> 뜻 : 4.568 × 10^-8
print(f"실패 확률: {prob:.4f}")  # 결과: 0.000
#.3e : 지수 표기법, 소수점 이하 3자리
#.4f : 고정 소수점, 소수점 이하 4자리
```

```python
user_input = input("아무 숫자 또는 단어를 입력하세요: ")
print(user_input)
print(type(user_input)) #input으로 받은 데이터는 항상 문자열(str) 자료형으로 저장된다.

#split은 공백을 기준으로 입력값을 변수 두 개에 저장한다
one,two = input("두 단어를 입력하세요").split()
print(one); print(two)
```

```python
#문자열 리스트를 계산하기 위해 map을 사용하여 정수&실수형으로 변환하기
a, b = map(float,input("실수 두 개를 입력하세요: ").split())
print(a + b)

x, y = map(int, input("정수 두 개를 더하기를 사용하여 입력하세요: ").split('+'))
print(x + y)
```

```python
# 1. 사용자로부터 정수 세 개를 공백으로 구분하여 입력받으시오.
a, b, c = map(int, input("정수 세 개를 입력하시오. : ").split())

# 2. 입력받은 세 정수의 합계를 계산하여 출력하시오.
print(a+b+c)

# 3. 입력받은 첫 번째 숫자에서 두 번째 숫자를 곱하고, 세 번째 숫자를 나눈 결과를 출력하시오.
print(a*b/c)
```

```python
# sep은 구분자라는 의미이며, 출력되는 값들 사이의 문자를 원하는 문자나 기호로 지정하여 조절할 수 있다.
print(1, 2, 3)
print(4, 5, 6, sep='+')
print('=')
print(4+5+6)
print(2026, 4, 17, sep='-')

print(1, end=' ')
print(2, end=' ')
print(3)
print('안녕하세요', end=' 정말 ')
print('만나서', end=' 반갑습니다!')
```

```python
# 1. 저장된 변수 값을 활용하여 year/month/day hour:minute:second 형태로 출력하시오.
year, month, day, = 2026, 4, 27
hour, minute, second = 16, 35, 2

print(year, month, day, sep='/', end=' ')
print(hour, minute, second, sep=':')

# 2. 제어 문자와 sep 옵션을 활용하여 아래 형태의 디렉터리 경로를 단 한 번의 print 함수로 출력하시오.
print('C:', 'Users', 'Admin', sep='\n')
```

```python
# 값을 비교할 때는 == 연산자를 사용, 객체 자체가 동일한지 비교할 때는 is 연산자를 사용
print(1.0 == 1)
print(1.0 is 1)

#bool 함수는 "정수 0, 실수 0.0, 아무 글자도 없는 빈 문자열"은 모두 거짓인 False로 변환하고,
#그 외의 0이 아닌 숫자와 글자는 모두 참인 True로 변환한다.
print(bool(3.14))
print(bool(0))
print(bool('은혜'))
print(bool())
```
