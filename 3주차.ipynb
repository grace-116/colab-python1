```python
# 1부터 10까지의 숫자 중에서 짝수만 골라 10을 곱한 리스트 만들기

# 기존의 일반적인 방식
standard_list = []
for num in range(1, 11):
    if num % 2 == 0:
        standard_list.append(num * 10)
print(standard_list)

# 리스트 내포를 사용한 방식 (위 4줄의 코드를 단 한 줄로 압축)
comprehension_list = [num * 10 for num in range(1, 11) if num % 2 == 0] # 구조: [표현식 for 변수 in 반복가능객체 if 조건식]
print(comprehension_list)
```

```python
# 중간 점검 종합 퀴즈: 공정 장비 센서 데이터 분석
'''
각 튜플은 `(장비 ID, 측정 온도, 상태)`의 정보를 담고 있다.
이 데이터를 분석하여 상태가 "CRITICAL"인 경고가 각 장비(Chamber)별로 몇 번이나 발생했는지 딕셔너리 형태로 집계하는 코드를 작성하시오.

**요구 사항:**
1. 결과를 저장할 빈 딕셔너리 `critical_counts`를 생성하시오.
2. `for` 반복문을 사용하여 `sensor_data` 리스트의 요소를 하나씩 순회하시오.
3. `if` 조건문과 인덱싱을 사용하여 현재 순회 중인 튜플의 상태 데이터가 "CRITICAL"인지 확인하시오.
4. 상태가 "CRITICAL"일 경우, 딕셔너리 내부에 해당 장비 ID를 나타내는 키가 이미 존재하는지 `in` 연산자로 검사하시오.
   - 키가 존재한다면 기존 값에 1을 더하십시오.
   - 키가 존재하지 않는다면 새로운 키를 만들고 값으로 1을 할당하십시오.
5. 반복문이 모두 종료된 후 `critical_counts` 딕셔너리를 출력하십시오.
'''

sensor_data = [
    ("Chamber_A", 25.5, "NORMAL"),
    ("Chamber_B", 27.1, "WARNING"),
    ("Chamber_A", 28.5, "CRITICAL"),
    ("Chamber_C", 24.8, "NORMAL"),
    ("Chamber_B", 29.0, "CRITICAL"),
    ("Chamber_A", 26.0, "NORMAL"),
    ("Chamber_C", 30.1, "CRITICAL"),
    ("Chamber_A", 29.5, "CRITICAL")
]

# 빈 딕셔너리 생성
critical_counts = {}

# 데이터 집계 로직
for data in sensor_data:
    if data[2] == "CRITICAL":
        if data[0] in critical_counts:
            critical_counts[data[0]] = critical_counts[data[0]] + 1
        else:
            critical_counts[data[0]] = 1

print(critical_counts)
```

```python
#초기 금액 10,000원이 들어있는 교통카드가 있습니다. 이 교통카드를 사용하여 요금이 1,350원인 버스를 타려고 합니다.
#`while` 반복문을 사용하여 현재 잔액을 출력하는 코드를 작성하십시오. 잔액이 부족해지면 반복문을 종료해야 합니다.

balance = 10000
fare = 1350

while balance >= fare:
  balance -= fare
  print("현재잔액:",balance)
```
